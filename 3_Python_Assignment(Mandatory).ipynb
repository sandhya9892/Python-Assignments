## Q5: Find the closest points
consider you have given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q)

your task is to find 5 closest points(based on cosine distance) in S from P

cosine distance between two points (x,y) and (p,q) is defind as  cos1((xp+yq)/(x2+y2)(p2+q2))
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]

P= (3,-4)



Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples
# you can free to change all these codes/structure


# here S is list of tuples and P is a tuple ot len=2
def closest_points_to_p(S, P):
    # write your code here
    Closest_pts = []
    
    for i in range(len(S)):
        Distance = math.acos(((S[i][0]*P[0]) + (S[i][1]*P[1])) / ((math.sqrt((S[i][0])**2)+ (math.sqrt((S[i][1])**2))) * \
                                                       (math.sqrt((P[0])**2)+ (math.sqrt((P[1])**2)))))
        #print(i,Distance)
        Closest_pts.append(Distance)
    #print("Closest_pts:",Closest_pts)
    
    closest_points_to_p = dict(zip(S,Closest_pts))
    #print("dict",dictionary)
    
    return closest_points_to_p  

   
#S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)]
S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]

#P=(p,q)
P= (3,-4)

points = closest_points_to_p(S, P)
#print the returned values
print("Top 5 Closest point to P are : ") 
for key, value in sorted(points.items(), key=lambda item: item[1])[:5]:
    print(key)
```

Ex 2: 40, _, _, _, 60 ==> (60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5 ==> 20, 20, 20, 20, 20 i.e. the sum of (60+40) is distributed qually to all 5 places

Ex 3: 80, _, _, _, _  ==> 80/5,80/5,80/5,80/5,80/5 ==> 16, 16, 16, 16, 16 i.e. the 80 is distributed qually to all 5 missing values that are right to it

Ex 4: _, _, 30, _, _, _, 50, _, _  

==> we will fill the missing values from left to right 

    a. first we will distribute the 30 to left two missing values (10, 10, 10, _, _, _, 50, _, _)
    
    b. now distribute the sum (10+50) missing values in between (10, 10, 12, 12, 12, 12, 12, _, _) 
    
    c. now we will distribute 12 to right side missing values (10, 10, 12, 12, 12, 12, 4, 4, 4)
    
    for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _" you need fill the missing values
Q: your program reads a string like ex: "_, _, x, _, _, _" and returns the filled sequence

Ex:

Input1: $"_,_,_,24"$

Output1: 6,6,6,6

Input2: $"40,_,_,_,60"$

Output2: 20,20,20,20,20

Input3: $"80,_,_,_,_"$

Output3: 16,16,16,16,16

Input4: $"_,_,30,_,_,_,50,_,_"$

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings


# you can free to change all these codes/structure
def curve_smoothing(string):
    """
    Program of Filling the missing values in the specified formate
    """
    #print(string)
    split_str = string.split(',')
    print("\nOriginal string :",split_str)
    Length = len(split_str)
    #print("Length ",Length)
    
    Digit_index = []
        
    for i in range(len(split_str)):
        if split_str[i] != '_': 
            Digit_index.append(i) 
                        
    Digit_index.append(len(split_str) - 1)  
    #print("Digit_index",Digit_index)
        
    start = 0
    for i in Digit_index:
        cum_sum = int(split_str[i]) if split_str[i] != '_' else 0
        cum_sum += int(split_str[start]) if split_str[start] != '_' and start != i else 0
                
        Replaced_val = cum_sum // ( i - start + 1)
        #print("Replaced_val",Replaced_val)
        
        split_str = [Replaced_val if start <= x <= i else split_str[x] for x in range(len(split_str))]
        start = i
        
    return split_str

str1 = "_,_,_,24"
smoothed_values= curve_smoothing(str1)
print("String 1 Output:",smoothed_values)

str2 = "40,_,_,_,60"
smoothed_values= curve_smoothing(str2)
print("String 2 Output:",smoothed_values)

str3 = "80,_,_,_,_"
smoothed_values= curve_smoothing(str3)
print("String 3 Output:",smoothed_values)

str4=  "_,_,30,_,_,_,50,_,_"
smoothed_values= curve_smoothing(str4)
print("String 3 Output:",smoothed_values)
print(curve_smoothing.__doc__)
```

[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]

$1/8((1log10(0.4)+0log10(0.6))+(0log10(0.5)+1log10(0.5))+...+(1log10(0.8)+0log10(0.2)))$

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings
import math as m

# you can free to change all these codes/structure
def compute_log_loss(A):
    length = len(A)
    sum = 0
    for Row in A:
        sum += (Row[0] * m.log(Row[1], 10)) + ((1-Row[0]) * m.log(1 - Row[1], 10))
       
    loss = ((-1) / length) * sum
           
    return loss

A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print("Output : ",loss)
```
